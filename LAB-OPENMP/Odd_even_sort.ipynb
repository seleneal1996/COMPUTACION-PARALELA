```python
import matplotlib.pyplot as plt
from math import log as ln
from matplotlib.pyplot import figure
x=[]
y=[]
y1=[]
y2=[]
tam_vec = [100,200,400,600,800,1000,1200,1400]

for i in tam_vec:
  x.append(i)
for i in [  4.100860e-04 ,8.434740e-04 ,2.371052e-02,3.187885e-02 ,9.571026e-02 ,1.737769e-01 ,3.989683e-01 ,7.550444e-01]: y.append(i)
for i in [3.305380e-04,8.851860e-04 ,1.771022e-03 ,5.656107e-02 ,1.008414e-01  ,2.075073e-01 ,3.165013e-01 ,7.567078e-01 ]: y1.append(i)
for i in tam_vec:
  y2.append(ln(i)*i)

plt.plot(x,y,color ='b',label='First OpenMP Odd-Even Sort')
plt.plot(x,y1,color ='y',label='Second OpenMP Odd-Even Sort')

plt.grid(True)
plt.legend()
plt.xlabel('n')
plt.ylabel('times')
```
